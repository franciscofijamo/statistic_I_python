## Grafico de Abelhas

Ao contratrio do histograma, ele apresenta os dados individualmente representados. Os Histogramas em cada bin apresentam  uma collecao de dados, o que pode trazer algum vieis de interpretacao.
Para trabalhar com o grafico de abelha, voce precisa ter o seu dataframe bem organizado, onde cada coluna representa uma feature, e cada linha uma observacao.

```
sns.swarmplot( x= 'state', y = 'dem_share', data=df)
plt.xlabel('state')
plt.ylabel('percent of vote for Obama')

plt.show()
```

# Plote todos os seus dados: ECDFs
 Em casos que o numero de observacoes for muito superior, o grafico de enchame de abelhas nao e a melhor alternativa, pois os dados ficam sobrepostos, e fica dificil ter uma nocao de quantidades. Para isso podemos usao o ECDFs (Empirical cumulative destribution function) Onde o eixo X representa a percentagem do que pretendemos contar. E o eixo Y a percentagem de concentracao.



```
x = np.sort(df['dem_share'])
y = np.arange(1, len(x)+1) / len(x)
plt.plot(x, y, marker = '.', linestyle = 'none')

# rotulando os axes
plt.xlabel( 'percentagem do voto para Obama')
plt.ylabel( 'ECDF pessoas contadas')

plt.show()
# usamos margins para garantir que nenhum ponto dos dados seja executado na lateral da rea de plot
```

# Media, Mediana, Percentil,Outliers e Boxplot

```
# Media
de_mean = np.mean(df.dem_votes)

# Percentil
df_percentile = np.percentile(df.dem_votes,[25, 50, 75, 99]) 

print(" Medidas de dem_votes, Median : {}, Percentile{}".format(de_mean,df_percentile))
```

```
# Box plot

sns.boxenplot( x= 'state', y = 'dem_share', data=df)
plt.xlabel('state')
plt.ylabel('percent of vote for Obama')

plt.show()
```
